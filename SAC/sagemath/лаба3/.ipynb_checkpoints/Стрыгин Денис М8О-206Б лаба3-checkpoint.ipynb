<h1>Падение тела брошенного под углом к горизонту</h1>

```python
var("p1, p2, p3, p4")

t = np.linspace(0,7,300)
alpha = np.pi/4 
v0 = 10.0 
y0 = [0, v0*np.cos(alpha), 0, v0*np.sin(alpha)]
k = 0.01
G = 9.81 
f = [p2, -k*p2*sqrt(p2**2+p4**2), p4, -k*p2*sqrt(p2**2+p4**2)-G]
sol = desolve_odeint(f, y0, t, dvars=[p1, p2, p3, p4])
Y = np.array(sol)
plot_arr = np.vstack((Y[:, 0], Y[:, 2])).T
g = line(plot_arr)
g.show(xmin=0, xmax=10, ymin=0, ymax=4, aspect_ratio=1)
```
