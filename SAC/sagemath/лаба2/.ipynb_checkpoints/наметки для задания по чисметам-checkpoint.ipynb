```python
def interaction_function(_step):
    return plot(sin(2*pi*x), (x, 0, 10+_step))
```

```python
interaction_function(0.1)
```

```python
@interact
def _(step=(0.1, 10, 0.5)):
    show(interaction_function(_step=step))
```

```python
var("x")
# процесс: каждую итерацию получаем некоторое приближение к искомому значению корня
# каждый следующий X получается путем расчета значения функции phi(x)
```

$$x_{n+1} = \phi(x)$$
$$\phi(x) = \frac{f(x)}{f'(x)}$$
$$x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)}$$

```python
f(x) = cos(x) - x**3
show(f)
```

```python
df = f.derivative(1)
```

```python
show(df)
```

```python
def generate_point_positions(_quo: int):
    # функция генерирует массив кадров
    # т.е. возвращает массив x, y для нашей красной точки,
    # которая показывает текузее положение найденного приближения к корню
    x_n = a
    res_out = []
    for i in range(_quo):
        x_n = x_n - f(x_n)/df(x_n)
        res_out.append([x_n, f(x_n)])
    return res_out
```

```python
a = 0.4
b = 1.1
```

```python
plot(f, (x, a, b))
```

```python
# Сказано: жлемент Interact(ползунок) должен варьировать значение параметра n
# где n - номер шага приближения
# x0 = a  # x0 - есть начальное значение x, начиная с которого запускаем итерационный процесс
# т.е. начинаем искать, делать пристрелку
```

```python
quo_iterations = 10
frames = generate_point_positions(_quo=quo_iterations)
```

```python
def interaction_function(_n):
    return plot(f, (x, a, b)) + point(frames[_n], color="red", size=30, zorder=10)
```

```python
@interact
def _(n=(1, quo_iterations, 1)):
    show(interaction_function(_n=n))
```
