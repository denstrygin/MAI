```python
plot.options["xmin"] = -50
plot.options["xmax"] = 50
plot.options["ymin"] = -55
plot.options["ymax"] = 45

func = 3*cos(x)*x^3-7*sin(x)*x^2+15*x-7
plot(func, title="${}$".format(latex(func)))
```

```python
def newton(f, a, b, EPS):
    x1 = a
    x2 = (a + b) / 2
    
    df = f.derivative()

    while abs(x2 - x1) >= EPS:
        x1 = x2
        x2 = x1 - f(x=x1)/df(x=x1)

    return x1
```

```python
EPS = 1e-5
a = 1
b = 1.7

x_1 = newton(func, a, b, EPS)

p1 = plot(func,(0,2), title="${}$".format(func))
p1 += point((x_1, 0), color="red", size=30, zorder=20)
p1 += text(r"$x_1$", (x_1, 5), color="black", fontsize=14)

p1.show()
```

```python
# visualization newton

x, y = var("x,y")

x1 = a
x2 = (a + b)/2

p2 = plot(func,(0,2), title="${}$".format(func))

df = func.derivative()

while abs(x2 - x1) >= EPS:
    x1 = x2
    x2 = x1 - func(x=x1) / df(x=x1)
    l = df(x1)*(x - x1) + func(x=x1)
#     p2 += plot(l, color="red")
    (p2 + plot(l, color="red") + text((r"$y={}$" + "\n" + r"$x_1={}, x_2={}$").format(latex(l), x1, x2), (1.2, -15), fontsize=10, color="black")).show(xmin=1/2, xmax=2, ymin=-20, ymax=20)    

```
